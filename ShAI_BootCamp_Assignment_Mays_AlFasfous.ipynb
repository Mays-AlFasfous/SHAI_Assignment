**1- Basic Data Exploration: Identify the number of rows and columns in the dataset, determine the data types of each column, and check for missing values in each column.**

```python
# Identify the number of rows and columns in the dataset
num_rows, num_columns = df.shape
print(f'Number of rows: {num_rows}')
print(f'Number of columns: {num_columns}')
```

**2- Descriptive Statistics: Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.**

```python
# Calculate basic statistics
basic_stats = df['TotalPay'].describe()
print('Basic Statistics for Salary:')
print(basic_stats) #Note: 50% is equal median
```

```python
# Calculate basic statistics mean, median, mode, minimum, and maximum salary
mean_salary = np.mean(df['TotalPay'])
median_salary = np.median(df['TotalPay'])
mode_salary = df['TotalPay'].mode()[0]
min_salary = np.min(df['TotalPay'])
max_salary = np.max(df['TotalPay'])

print(f'Mean Salary: {mean_salary}')
print(f'Median Salary: {median_salary}')
print(f'Mode Salary: {mode_salary}')
print(f'Minimum Salary: {min_salary}')
print(f'Maximum Salary: {max_salary}')
```

```python
# Calculate range of salaries
salary_range = df['TotalPay'].max() - df['TotalPay'].min()
print(f'Range of Salaries: {salary_range}')
```

```python
# Find the standard deviation of salaries
salary_std_dev = df['TotalPay'].std()
print(f'Standard Deviation of Salaries: {salary_std_dev}')
```

**3- Data Cleaning: Handle missing data by suitable method with explain why you use it.**

```python
# Handle missing data
# Impute missing values in numerical columns with mean
numerical_columns = ['BasePay', 'OvertimePay', 'OtherPay']
df[numerical_columns] = df[numerical_columns].apply(lambda x: x.fillna(x.mean()))

# Impute missing values in 'Benefits' with 0
df['Benefits'].fillna(0, inplace=True)  # Imputing with 0 assumes that missing values in the 'Benefits' column indicate no benefits.

# Drop 'Benefits', 'Notes', and 'Status'
df.drop(['Notes', 'Status'], axis=1, inplace=True)
```

 **- explain why you use it:**

1- BasePay, OvertimePay, OtherPay:

Handling Method: Impute with the mean.

Explanation: Since these columns represent numerical values related to pay, imputing with the mean is a common and simple approach that helps maintain the overall distribution of the salary-related data.

2- Benefits:

Handling Method: Impute with 0

Explanation: The 'Benefits' column has a large number of missing values (36,163 out of 148,654), and imputing with the mean might introduce bias. So I could impute with 0 if I assume that missing values indicate no benefits.

3- Notes, Status:

Handling Method: Drop these columns.

Explanation: The 'Notes' and 'Status' columns have no non-null values; every entry in these columns is missing. As a result, these columns do not contain any meaningful information for analysis.Dropping columns with no variability simplifies the dataset, making it more efficient for analysis.

**4- Basic Data Visualization: Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.**

```python
import matplotlib.pyplot as plt
```

```python
# Create a histogram to visualize the distribution of salaries
plt.hist(df['TotalPay'], bins=10, edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a bar chart to visualize the distribution of salaries
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=20, edgecolor='black', color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Assuming 'JobTitle' contains the department information
department_counts = df['JobTitle'].value_counts().head() # I visualized only 5 'Job Title' to be more clear

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Employees in Different Departments')
plt.legend(title='Job Titles', loc='upper right', bbox_to_anchor=(1.9, 1))

plt.show()

# I did not assume the agency is same of the department because it contains only one value
```

**5- Grouped Analysis: Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.**

```python
# Group the data by one or more columns and calculate summary statistics for each group
grouped_data = df.groupby(['JobTitle', 'Year'])['TotalPay'].agg(['mean', 'median', 'min', 'max']).sort_values(by='Year')

print(grouped_data)
```

```python
# Group the data by one or more columns and calculate summary statistics for each group
grouped_data = df.groupby('Year')['TotalPay'].agg(['mean', 'median', 'min', 'max']).sort_values(by='Year')

print(grouped_data)
```

```python
# Group the data by 'Year' and compare the average salary for each year
avg_salary_by_year = df.groupby('Year')['TotalPay'].mean().sort_values()
print('Average Salary by year:')
print(avg_salary_by_year)
```

```python
# Group the data by department(JobTitle) and compare the average salary for each department(JobTitle)
avg_salary_by_department = df.groupby('JobTitle')['TotalPay'].mean().sort_values(ascending=False)
print('Average Salary by Department:')
print(avg_salary_by_department)
```

**6- Simple Correlation Analysis: Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.**

```python
# Identify correlation between salary and another numerical column ('BasePay')
correlation = df['TotalPay'].corr(df['BasePay'])
print(f'Correlation between TotalPay and BasePay: {correlation}')
```

```python
# Plot a scatter plot to visualize the relationship
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BasePay', y='TotalPay', data=df, alpha=0.9)
plt.title('Scatter Plot: TotalPay vs BasePay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Identify correlation between salary and another numerical column ('OvertimePay ')
correlation = df['TotalPay'].corr(df['OvertimePay'])
print(f'Correlation between TotalPay and OvertimePay: {correlation}')
```

```python
# Plot a scatter plot to visualize the relationship
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='OvertimePay', y='TotalPay', data=df, alpha=0.9)
plt.title('Scatter Plot: TotalPay vs OvertimePay')
plt.xlabel('OvertimePay')
plt.ylabel('TotalPay')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
